### Aim: To implement TFIDF model for document analysis

```python
# import statments
import numpy
import re
import nltk 
nltk.download('stopwords')
from nltk.corpus import stopwords 
stop_words = set(stopwords.words('english'))
def word_extraction(sentence):
    words = re.sub("[^\w]", " ",  sentence).split()
    cleaned_text = [w.lower() for w in words if w not in stop_words]
    return cleaned_text    

allsentences = ["Joe waited for the train", "The train was late", "Mary and Samantha took the bus", 
            "I looked for Mary and Samantha at the bus station", 
            "Mary and Samantha arrived at the bus station early but waited until noon for the bus"]

def cleaning(allsentences):
      clean_sen = []
      for sentence in allsentences:
        words = word_extraction(sentence)
        clean_sen.append(' '.join(words))
      return clean_sen

updated_sen = cleaning(allsentences)
```

```python
updated_sen
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfvectorizer = TfidfVectorizer()
```

```python
Y = tfidfvectorizer.fit_transform(updated_sen)
```
